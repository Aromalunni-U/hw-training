```python
import pandas as pd
```

```python
path = "/home/user/Desktop/hw-training/2025-08-25/hm/hm_20250825.csv"
df = pd.read_csv(path)
```

```python
df.columns
```

```python
df.drop(
    [
        "web-scraper-order", "pagination-selector",
        "web-scraper-start-url", "data-page-selector"
    ]
    ,axis=1,
    inplace=True
)
```

```python
df.rename(
    columns={
        "data-page-selector-href":"pdp_url",
        "image-0-src":"image",
        "text-1": "selling_price",
        "text-2": "product_name",
        "text-3": "product_description",
        "text-4": "size",
        "text-5": "color",
        "text-6": "review",
        },
    inplace=True
)
```

```python
df["review"] = df["review"].fillna("").astype(str)
df["review"] = df["review"].str.replace("Reviews", "", regex=False).str.replace("[", "", regex=False).str.replace("]", "", regex=False)


df["selling_price"] = df["selling_price"].apply(
    lambda x : str(x).replace("Rs.", "").replace(",","")
    if x else ""
)

df["size"] = df["size"].apply(
    lambda x: x.replace("\n", " ").split()
)

df["image"] = df["image"].apply(
    lambda x: x.split("?")[0] if x else ""
)
```

```python
df = df.fillna("")
df.to_csv("cleaned_hm.csv", index=False, na_rep="")
```
